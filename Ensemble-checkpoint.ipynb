```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor

import xgboost as xgb
```

```python
HVAC= pd.read_excel(r'../Data/HVAC_Energy.xlsx',index_col='Timestamp')

HVAC.dropna(inplace=True)


train_start = pd.to_datetime('2018-09-15 00:00:00')
train_end = pd.to_datetime('2018-10-14 23:50:00')
x_columns = ['Global UCF VTU 3-1 RM 300D CTL STPT','dayofweek','minuteofhour','hourofday']
x_train = HVAC.loc[train_start:train_end,x_columns]
y_train= HVAC.loc[train_start:train_end, 'Global UCF VerisH8186_5536BAD0 KW_Total']

test_start = pd.to_datetime('2018-10-15 00:00:00')
test_end = HVAC.index[-1]
x_test = HVAC.loc[test_start:test_end,x_columns]
y_test = HVAC.loc[test_start:test_end,'Global UCF VerisH8186_5536BAD0 KW_Total']
```

```python
# bagging
bg = BaggingRegressor().fit(x_train,y_train)
bg_pred = bg.predict(x_test)
```

```python
# RF
rf = RandomForestRegressor(max_depth=8,n_estimators=100).fit(x_train,y_train)
rf_pred = rf.predict(x_test)
```

```python
# GB
gbr = GradientBoostingRegressor().fit(x_train,y_train)
gbr_pred = gbr.predict(x_test)
```

```python
xgb_model = xgb.XGBRegressor().fit(x_train,y_train)
xgb_pred = xgb_model.predict(x_test)
```

```python
# plot
plt.figure(figsize=(16,9))
plt.plot(x_test.index, y_test,x_test.index, bg_pred,x_test.index,rf_pred,x_test.index,gbr_pred,x_test.index,xgb_pred)
plt.legend(['True','Bagging','RF','GB'])
plt.title('Ensemble Method to Predict HVAC Energy Consumption')
```
