```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score


plt.rcParams['figure.figsize']=(16,9)
#%matplotlib notebook
```

```python
AHU = pd.read_excel(r'../Data/Global_Clean.xlsx','AHU',index_col='Timestamp')
```

```python
AHU.shape[1]
```

```python
AHU.plot(subplots=True,figsize=(16,40))
```

```python
AHU.iloc[300:900,:].plot(subplots=True,figsize=(16,40))
```

```python
corr = AHU.corr()
sns.set(style="white")
cmap = sns.diverging_palette(220, 10, as_cmap=True)
fig = plt.figure(figsize=(16,9))
sns.heatmap(corr,cmap=cmap,vmin=-1,vmax=1,center=0)
```

```python
AHU.columns
```

```python
AHU.fillna(method='bfill',inplace=True)
features = ['Global UCF AHU-3 OADP',
       'Global UCF AHU-3 SAT', 'Global UCF AHU-3 CCV', 'Global UCF AHU-3 CCT',
       'Global UCF AHU-3 RAH', 'Global UCF AHU-3 RAT']
x_train = AHU.loc['2018-09-15 00:00:00':pd.to_datetime('2018-10-15 16:30:00')-pd.to_timedelta(72,unit='h'), features]
y_train = AHU.loc['2018-09-15 00:00:00':pd.to_datetime('2018-10-15 16:30:00')-pd.to_timedelta(72,unit='h'), 'Global UCF AHU-3 SAF']
```

```python
x_test = AHU.loc['2018-10-12 16:40:00':'2018-10-15 16:30:00',features]
y_test = AHU.loc['2018-10-12 16:40:00':'2018-10-15 16:30:00','Global UCF AHU-3 SAF']
```

```python
clf = LogisticRegression()
clf.fit(x_train,y_train)
```

```python
y_estimate = clf.predict(x_test)
y_estimate
```

```python
y_prob = clf.predict_proba(x_test)
y_prob
```

```python
fig = plt.figure(figsize=(16,9))
plt.subplot(111)
plt.plot(y_test.index,y_test,'r*',y_test.index,y_estimate,'-')
plt.show()
```

```python
accuracy_score(y_test,y_estimate,normalize=False)
```

```python
y_test.shape
```

```python
# Naive Bayes
gnb = GaussianNB()
gnb.fit(x_train,y_train)
y_estimate_nb = gnb.predict(x_test)
```

```python
fig = plt.figure(figsize=(16,9))
plt.subplot(111)
plt.plot(y_test.index,y_test,'r*',y_test.index,y_estimate_nb,'-')
plt.show()
```

```python
accuracy_score(y_test,y_estimate_nb,normalize=False)
```

```python
# Naive Bayes
mnb = MultinomialNB()
mnb.fit(x_train,y_train)
y_estimate_mb = mnb.predict(x_test)
```

```python
fig = plt.figure(figsize=(16,9))
plt.subplot(111)
plt.plot(y_test.index,y_test,'r*',y_test.index,y_estimate_mb,'-')
plt.show()
```

```python
accuracy_score(y_test,y_estimate_mb,normalize=False)
```
