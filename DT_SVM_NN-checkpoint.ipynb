```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


from sklearn import tree
from sklearn import svm
```

```python
HVAC= pd.read_excel(r'../Data/HVAC_Energy.xlsx',index_col='Timestamp')

```

```python
HVAC.dropna(inplace=True)
```

```python

train_start = pd.to_datetime('2018-09-15 00:00:00')
train_end = pd.to_datetime('2018-10-14 23:50:00')
x_columns = ['Global UCF VTU 3-1 RM 300D CTL STPT','dayofweek','minuteofhour','hourofday']
x_train = HVAC.loc[train_start:train_end,x_columns]
y_train= HVAC.loc[train_start:train_end, 'Global UCF AHU-3 SAF']

test_start = pd.to_datetime('2018-10-15 00:00:00')
test_end = HVAC.index[-1]
x_test = HVAC.loc[test_start:test_end,x_columns]
y_test = HVAC.loc[test_start:test_end,'Global UCF AHU-3 SAF']
```

```python
DT = tree.DecisionTreeClassifier()
clf = DT. fit(x_train,y_train)
```

```python
dot_data=tree.export_graphviz(clf,out_file=None)
graph = graphviz.Source(dot_data)
graph
```

```python
y_pred = clf.predict(x_test)
fig = plt.figure(figsize=(16,9))
plt.plot(x_train.tail(100).index,y_train.tail(100))
plt.plot(x_test.index,y_test,'o',x_test.index,y_pred,'-')
plt.xlabel('Time')
plt.ylabel('HVAC Fan Status')
plt.legend(['real','predict'])
```
