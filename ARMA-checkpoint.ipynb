```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX
import statsmodels.graphics.tsaplots as tsaplots
from statsmodels.graphics.gofplots import qqplot
import matplotlib.dates as mdates
import scipy.stats as stats

#%matplotlib inline
plt.rcParams['figure.figsize'] = (16,9)
```

```python
power = pd.read_excel(r'..\Data\Global_Clean.xlsx',sheet_name='power',index_col='Timestamp')
```

```python
plt.plot(power.head(200))
```

```python
power_diff = power.diff(periods=1)
plt.plot(power_diff.head(200))
```

```python
fig = plt.figure()
ax = plt.subplot(211)
tsaplots.plot_acf(power,lags=100, ax=ax)

ax = plt.subplot(212)
tsaplots.plot_pacf(power,lags=100, ax=ax)

plt.show()
```

```python
split_time = pd.to_datetime('2018-10-14 00:00:00')
train_data = power.loc[:split_time-pd.to_timedelta(15,unit='m')]
test_data = power.loc[split_time-pd.to_timedelta(2,unit='d'):]
```

```python
arima_model = SARIMAX(train_data,order=(2,1,2),freq='15min')
arima_result = arima_model.fit()
arima_result.summary()
```

```python
# model diagonostics: residuals, qqplot, histogram, and ACF

fig = plt.figure(figsize=(16,9))
ax = plt.subplot(221)
plt.plot(arima_result.resid)
ax.xaxis.set_major_formatter(mdates.DateFormatter("%m-%d"))


ax = plt.subplot(222)
plt.hist(arima_result.resid)

ax = plt.subplot(223)
qqplot(arima_result.resid,line='s',ax=ax)

ax = plt.subplot(224)
tsaplots.plot_acf(arima_result.resid, lags=200,ax=ax)

plt.show()
```

```python
# seasonal component
sarima_model = SARIMAX(train_data,order=(2,1,2),seasonal_order=(1,1,0,96),freq='15min')
sarima_result = sarima_model.fit()
sarima_result.summary()
```
